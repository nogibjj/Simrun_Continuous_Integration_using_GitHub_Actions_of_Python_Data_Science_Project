#### Descriptive Statistics on Diabetic Patients at least 21 years old of Pima Indian heritage: Data provided by National Institute of Diabetes and Digestive and Kidney Diseases. 
   

I first wanted to do a statistical analysis of the data on the three predictor variables I choose.  I investigated three different medical predictors of diabetes: Glucose, Insulin, BMI. In effort to understand if a patient had diabetes or not (1- had diabetes, 0 - does not have diabetes)

```python
import lib
import pandas as pd
```

```python
# importing functions from lib.py
data = pd.read_csv('../data/diabetes.csv')
column_g = "Glucose"
column_i = "Insulin"
column_b = "BMI"

print("Column Name: ", column_g)
print('Maximum: ', lib.maximum(data,column_g))
print('Minimum: ', lib.minimum(data,column_g))
print('Mean: ', lib.get_mean(data,column_g))
print('Median: ', lib.get_median(data,column_g))
print('Standard Deviation: ', lib.get_std_dev(data,column_g))
print("\n")
print("Column Name: ", column_i)
print('Maximum: ', lib.maximum(data,column_i))
print('Minimum: ', lib.minimum(data,column_i))
print('Mean: ', lib.get_mean(data,column_i))
print('Median: ', lib.get_median(data,column_i))
print('Standard Deviation: ', lib.get_std_dev(data,column_i))
print("\n")
print("Column Name: ", column_b)
print('Maximum: ', lib.maximum(data,column_b))
print('Minimum: ', lib.minimum(data,column_b))
print('Mean: ', lib.get_mean(data,column_b))
print('Median: ', lib.get_median(data,column_b))
print('Standard Deviation: ', lib.get_std_dev(data,column_b))

```
